```python
from sympy import symbols, diff, Interval, Float, N, Mul
from sympy.plotting import plot
```

```python
x = symbols("x")
f = x / (x**2 - 5*x + 10)
f
```

```python
p = plot(f)
```

```python
from sympy.calculus.util import minimum, maximum
N(maximum(f, x))
```

```python
f_prime = diff(f, x)
f_prime
```

```python
from sympy import symbols, diff, solve
# Находим критические точки, где производная равна нулю
critical_points = solve(f_prime, x)

# Проверяем тип точек (максимум, минимум или точка перегиба)
for point in critical_points:
    second_derivative = diff(f_prime, x).subs(x, point)
    if second_derivative < 0:
        print(f"Функция принимает максимум в точке x = {N(point)}")
```
